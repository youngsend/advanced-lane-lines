```python
from camera_calibration import *
import matplotlib.pyplot as plt
```

```python
camera_calibration = CameraCalibration('camera_cal/')
```

```python
img = cv2.imread('camera_cal/calibration1.jpg')
```

```python
dst = camera_calibration.undistort(img)
```

```python
def show_two_images(img, dst):
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 9))
    f.tight_layout()
    ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    ax1.set_title('Original Image', fontsize=50)
    ax2.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
    ax2.set_title('Undistorted and Warped Image', fontsize=50)
    plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```

```python
show_two_images(img, dst)
```

```python
cv2.imwrite('output_images/undistorted_chessboard1.jpg', dst)
```

```python
img2 = cv2.imread('test_images/test1.jpg')
dst2 = camera_calibration.undistort(img2)
```

```python
show_two_images(img2, dst2)
```

```python
cv2.imwrite('output_images/undistorted_test1.jpg', dst2)
```
